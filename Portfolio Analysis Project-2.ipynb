```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
from scipy import stats
import seaborn as sns
import yfinance as yf
import datetime as dt
```

```python
#beta against SPY
tickers =["msft","aapl","amzn","jpm","tsla","tstrx","ms"]
weight= [14.285,14.285,14.285,14.285,14.285,14.285,14.285]
price_data = yf.download(tickers,
                            start = '2022-02-20',
                               end = '2023-02-20')
price_data = price_data['Adj Close']
ret_data = price_data.pct_change()[1:]
port_ret = (ret_data * weight).sum(axis = 1)
benchmark_price = yf.download('SPY',
                               start = '2022-02-20',
                               end = '2023-02-20')
                               
benchmark_ret = benchmark_price["Adj Close"].pct_change()[1:]
ret_data = price_data.pct_change()[1:]
port_ret = (ret_data * weight).sum(axis = 1)
benchmark_price = yf.download('SPY',
                            start = '2022-02-20',
                               end = '2023-02-20')
sns.regplot(benchmark_ret.values,
port_ret.values)
plt.xlabel("Benchmark Returns")
plt.ylabel("Portfolio Returns")
plt.title("Portfolio Returns vs Benchmark Returns")
plt.show()
```

```python
(beta, alpha) = stats.linregress(benchmark_ret.values,
                port_ret.values)[0:2]
                
print("The portfolio beta against the s&p is", round(beta, 4)/100)
```
